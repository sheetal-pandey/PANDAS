```python
d4=pd.DataFrame({'A':[1,2,np.nan],
                'B':[33,np.nan,np.nan],
                'C':[2,3,4]})
```

```python
d4
```

```python
d4.dropna(how='all')
```

```python
d4.dropna(axis=1)
```

```python
d4.dropna(thresh=2)
```

```python
d4.fillna(value=50)
```

```python
d4.fillna(value='abc')
```

```python
d4.fillna({"A":150,"B":200,"C":300})
```

```python
d4.fillna(method='ffill')
```

```python
d4.mean()
```

```python
d4=pd.DataFrame({'A':[1,2,np.nan],
                'B':[33,np.nan,np.nan],
                'C':[2,3,4]})
d4.fillna(d4.mean())
```
